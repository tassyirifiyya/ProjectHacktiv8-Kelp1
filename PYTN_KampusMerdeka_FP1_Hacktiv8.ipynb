### plot_line_and_bar

Sekarang kita akan membut funsi plot_line_and_bar. Disni plotnya terdiri dari 3 parameter diantaranya kolom1, kolom2, dan ninety. ninety ini digunakan untuk rotasi penamaan pada nama sumbu X. Jika namanya panjang nantinya bisa kita rotasikan 90 derajat

```python
def plot_line_and_bar(groupby_column,ninety): # Definisikan fungsi plot_line_and_bar dengan parameter groupby_column dan ninety
    df1 = df_datacleansing_1.groupby(groupby_column).size().reset_index(name="counts")  # Mengelompokkan data dan menghitung jumlah data dalam setiap grup
    x = df1[groupby_column]  # Menyimpan data kolom groupby_column ke dalam variabel x
    y = df1["counts"]  # Menyimpan data kolom counts ke dalam variabel y
    num_colors = len(df1)  # Menghitung jumlah warna yang dibutuhkan untuk plot
    colormap = plt.cm.get_cmap('tab20', num_colors)  # Mengambil colormap dengan jumlah warna yang sesuai
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(20, 10))  # Membuat objek plot dengan 2 subplot, serta mengatur ukuran figure menjadi 20x10 inches
   
    # Plot Line
    for i in range(num_colors):
        ax1.plot(x, y, marker='o', color=colormap(i))
    ax1.set_ylabel('Jumlah')
    if groupby_column == "month":
        ax1.set_xticks([11, 12])
    else:
        ax1.set_xlabel(groupby_column)
    ax1.set_title(f'Line Plot untuk {groupby_column}') 
    if ninety == 1:
        ax1.set_xticklabels(x, rotation=90)

    # Plot Bar
    bars = ax2.bar(x, y, color=colormap(np.arange(num_colors)))
    if groupby_column == "month":
        plt.xticks([11, 12])
    else:
        ax2.set_xlabel(groupby_column)
    ax2.set_ylabel('Jumlah')
    ax2.set_title(f'Bar Plot untuk {groupby_column}')
    if ninety == 1:
        ax2.set_xticklabels(x, rotation=90)
    ax2.legend(bars, df1[groupby_column], title=groupby_column, bbox_to_anchor=(1.05, 1), loc='upper left', fontsize=12) 

    #show data 
    a = input("Pilih untuk tampilan data [1] Semua data, [2] 5 Data terbesar, [3] Keduanya, [4] Tidak keduanya  : ")
    
    if a == "1":
        print(df1)
    elif a == "2":
        df_top_5 = df1.nlargest(5, 'counts')
        print(df_top_5)
    elif a == "3":
        print("All Data")
        print(df1)
        df_top_5 = df1.nlargest(5, 'counts')
        print(" TOP 5 ")
        print(df_top_5)
    else:
        print("Data disembunyikan")

    plt.show()
```

### plt_line_double

Untuk plt_line_double kita akan membandingkan 2 kolom, yang nantinya akan dijumlah dengan data yang ada. Dari fungsi ini kita akan bisa melihat apa saja yang dapat mempengaruhi persebaran datanya berupa line

```python
# Membuat fungsi plt_line_double dengan 2 argumen
def plt_line_double(groupby1, groupby2,ninety):
    # Melakukan groupby pada DataFrame df berdasarkan kolom groupby1 dan groupby2, dan menghitung jumlah
    df2 = df_datacleansing_1.groupby(by=[groupby1, groupby2]).size().reset_index(name="counts")
    
    # Mengambil nilai x dan y dari DataFrame df2
    x = df2[groupby1]
    y = df2["counts"]
    
    # Mengambil nilai warna dari DataFrame df2
    colors = df2[groupby2]
    
    # Membuat objek figure dan axis pada plot
    fig, ax = plt.subplots(figsize=(20,15))
    
    # Mengiterasi warna yang ada pada DataFrame df2
    for cab_type in df2[groupby2].unique():
        # Membuat plot garis dengan marker berbentuk 'o' untuk setiap warna yang berbeda
        ax.plot(x[colors == cab_type], y[colors == cab_type], marker='o', label=cab_type)

    # Menyertakan label pada sumbu x dan y, serta judul plot
    ax.set_xlabel(groupby1)
    ax.set_ylabel('Counts')
    ax.set_title('Line Chart')
    if ninety==1:
        ax1.set_xticklabels(x, rotation=90)
    
    # Menyertakan legenda pada plot
    ax.legend()
    
    # Menampilkan plot
    a = input("Pilih untuk tampilan data [1] Semua data, [2] 5 Data terbesar, [3] Keduanya, [4] Tidak keduanya  : ")
    if a=="1":
        print(df2)
    elif a=="2":
        df_top_5 = df2.nlargest(5, 'counts')
        print(df_top_5)
    elif a=="3":
        print("All Data")
        print(df1)
        df_top_5 = df2.nlargest(5, 'counts')
        print(" TOP 5 ")
        print(df_top_5)
    else:
        print("Data disembunyikan")
        
    plt.show()
```

### plt_bar_double

Untuk plt_bar_double kita akan membandingkan 2 kolom, yang nantinya akan dijumlah dengan data yang ada. Dari fungsi ini kita akan bisa melihat apa saja yang dapat mempengaruhi persebaran datanya berupa barchart

```python
def plt_bar_double(groupby1, groupby2):
    # Melakukan groupby pada DataFrame df berdasarkan kolom groupby1 dan groupby2, dan menghitung jumlah
    df2 = df_datacleansing_1.groupby(by=[groupby1, groupby2]).size().reset_index(name="counts")

    # Mengambil nilai x dan y dari DataFrame df2
    x = df2[groupby1]
    y = df2["counts"]

    # Mengambil nilai warna dari DataFrame df2
    colors = df2[groupby2]

    # Menentukan palet warna "tab20" dari matplotlib
    color_palette = plt.cm.get_cmap("tab20", len(df2[groupby2].unique()))

    # Membuat objek figure dan axis pada plot
    fig, ax = plt.subplots(figsize=(20,15))

    # Mengiterasi warna yang ada pada DataFrame df2
    for i, cab_type in enumerate(df2[groupby2].unique()):
        # Membuat plot bar dengan lebar 0.35 untuk setiap warna yang berbeda
        ax.bar(x[colors == cab_type], y[colors == cab_type], width=0.35, label=cab_type, color=color_palette(i))

    # Menyertakan label pada sumbu x dan y, serta judul plot
    ax.set_xlabel(groupby1)
    ax.set_ylabel('Counts')
    ax.set_title('Bar Chart')

    # Menyertakan legenda pada plot
    ax.legend()

    # Menampilkan plot
    # Menampilkan plot
    a = input("Pilih untuk tampilan data [1] Semua data, [2] 5 Data terbesar, [3] Keduanya, [4] Tidak keduanya  : ")
    
    if a=="1":
        print(df2)
    elif a=="2":
        df_top_5 = df2.nlargest(5, 'counts')
        print(df_top_5)
    elif a=="3":
        print("All Data")
        print(df1)
        df_top_5 = df2.nlargest(5, 'counts')
        print(" TOP 5 ")
        print(df_top_5)
    else:
        print("Data disembunyikan")
        
    plt.show()
```

```python
def one_hot_encoder(data, feature, keep_first=True):
    # Melakukan one-hot encoding pada fitur yang diberikan
    one_hot_cols = pd.get_dummies(data[feature])
    
    # Mengganti nama setiap kolom dalam DataFrame one_hot_cols
    for col in one_hot_cols.columns:
        one_hot_cols.rename({col:f'{feature}_'+col}, axis=1, inplace=True)
    
    # Menggabungkan DataFrame one_hot_cols dengan dataset data
    new_data = pd.concat([data, one_hot_cols], axis=1)
    # Menghapus fitur yang telah diencode
    new_data.drop(feature, axis=1, inplace=True)
    
    # Jika keep_first di set sebagai False, maka fitur pertama akan dihapus dari hasil encoding
    if keep_first == False:
        new_data = new_data.iloc[:, 1:]
    
    return new_data
```
